```python
import ipywidgets as widgets
from IPython.display import display
```

```python
grade_points = {
    "O": 10,
    "A+": 9,
    "A": 8,
    "B+": 7,
    "B": 6,
    "C+": 5,
    "C": 4,
    "U": 0
}

output = widgets.Output()
```

```python
mode_selector = widgets.RadioButtons(
    options=['GPA Calculator', 'CGPA Calculator'],
    description='Mode:',
    style={'description_width': 'initial'}
)
```

```python
subject_count = widgets.Dropdown(
    options=list(range(1, 51)),
    description='Subjects:'
)

generate_subjects_btn = widgets.Button(
    description='Generate Subjects',
    button_style='primary'
)

calculate_gpa_btn = widgets.Button(
    description='Calculate GPA',
    button_style='danger',
    disabled=True
)

gpa_inputs_box = widgets.VBox()
grade_inputs = []
credit_inputs = []

# ---------- CGPA MODE WIDGETS ----------
semester_count = widgets.Dropdown(
    options=list(range(1, 21)),
    description='Semesters:'
)

generate_sem_btn = widgets.Button(
    description='Generate Semesters',
    button_style='primary'
)

calculate_cgpa_btn = widgets.Button(
    description='Calculate CGPA',
    button_style='danger',
    disabled=True
)

cgpa_inputs_box = widgets.VBox()
semester_gpa_inputs = []

# ---------- GPA: Generate Subjects ----------
def generate_subjects(b):
    global grade_inputs, credit_inputs
    grade_inputs = []
    credit_inputs = []
    rows = []
    output.clear_output()
    
    for i in range(subject_count.value):
        grade = widgets.Dropdown(
            options=list(grade_points.keys()),
            description=f'Sub {i+1} Grade:',
            style={'button_color': '#0d6efd'}
        )
        credit = widgets.IntText(
            description='Credit:',
            min=0,
            max=5
        )
        grade_inputs.append(grade)
        credit_inputs.append(credit)
        rows.append(widgets.HBox([grade, credit]))
    
    gpa_inputs_box.children = rows
    calculate_gpa_btn.disabled = False
    calculate_gpa_btn.button_style = 'danger'

generate_subjects_btn.on_click(generate_subjects)

# ---------- GPA: Calculate ----------
def calculate_gpa(b):
    output.clear_output()
    
    total_points = 0.0
    total_credits = 0
    
    with output:
        for i in range(len(grade_inputs)):
            grade = grade_inputs[i].value
            credit = credit_inputs[i].value
            point = grade_points[grade]
            
            print(f"Subject {i+1}: Grade={grade}, Credit={credit}")
            
            if credit > 0 and grade != "U":
                total_points += point * credit
                total_credits += credit
        
        if total_credits == 0:
            print("\nGPA: 0.00")
            return
        
        gpa = float(f"{(total_points / total_credits):.2f}")
        print("\nTotal Credits Considered:", total_credits)
        print("GPA:", gpa)
    
    calculate_gpa_btn.button_style = 'success'

calculate_gpa_btn.on_click(calculate_gpa)

# ---------- CGPA: Generate Semesters ----------
def generate_semesters(b):
    global semester_gpa_inputs
    semester_gpa_inputs = []
    rows = []
    output.clear_output()
    
    for i in range(semester_count.value):
        gpa = widgets.FloatText(
            description=f'Sem {i+1} GPA:',
            min=0,
            max=10
        )
        semester_gpa_inputs.append(gpa)
        rows.append(gpa)
    
    cgpa_inputs_box.children = rows
    calculate_cgpa_btn.disabled = False
    calculate_cgpa_btn.button_style = 'danger'

generate_sem_btn.on_click(generate_semesters)
```

```python
def calculate_cgpa(b):
    output.clear_output()
    
    total = 0.0
    count = 0
    
    with output:
        for i, gpa_box in enumerate(semester_gpa_inputs):
            gpa = gpa_box.value
            print(f"Semester {i+1} GPA: {gpa}")
            total += gpa
            count += 1
        
        if count == 0:
            print("\nCGPA: 0.00")
            return
        
        cgpa = float(f"{(total / count):.2f}")
        print("\nCGPA:", cgpa)
    
    calculate_cgpa_btn.button_style = 'success'

calculate_cgpa_btn.on_click(calculate_cgpa)
```

```python
def switch_mode(change):
    output.clear_output()
    gpa_inputs_box.children = []
    cgpa_inputs_box.children = []
    
    if change['new'] == 'GPA Calculator':
        main_box.children = [
            mode_selector,
            subject_count,
            generate_subjects_btn,
            gpa_inputs_box,
            calculate_gpa_btn,
            output
        ]
    else:
        main_box.children = [
            mode_selector,
            semester_count,
            generate_sem_btn,
            cgpa_inputs_box,
            calculate_cgpa_btn,
            output
        ]

mode_selector.observe(switch_mode, names='value')
```

```python
main_box = widgets.VBox()
switch_mode({'new': 'GPA Calculator'})

display(main_box)
```
